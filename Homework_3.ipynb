```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.DataFrame.from_csv('C:\Users\khamz\Documents\us-marriages-divorces-1867-2014.csv')

class Marriage_Divorse:
    
    time_1 = data[(data.index >= '1867-01-01 00:00:00') & (data.index <= '1880-01-01 00:00:00')]
    
    time_2 = data[(data.index >= '1875-01-01 00:00:00') & (data.index <= '1881-01-01 00:00:00')]
    
    def __init__(self, data):
        self.data = data
    
d = Marriage_Divorse(data)

## Visualization
plt.figure(1, figsize=(15,10))
plt.subplot(2,2,1)
plt.title('Number of Marriages and Divorces \n per 1000 capita between 1867-1880', size = 15)
plt.ylabel('Number of Marriages and Divorces \n per 1000 capita', size = 12)
plt.xlabel('Years', size = 12)
M_1 = d.time_1['Marriages_per_1000'].plot()
D_1 = d.time_1['Divorces_per_1000'].plot()
plt.legend(['Marriages','Divorces'])

plt.subplot(2,2,2)
plt.title('Number of Marriages and Divorces \n per 1000 capita between 1875-1881', size = 15)
plt.ylabel('Number of Marriages and Divorces \n per 1000 capita', size = 12)
plt.xlabel('Years', size = 12)
M_2 = d.time_2['Marriages_per_1000'].plot()
D_2 = d.time_2['Divorces_per_1000'].plot()
plt.legend(['Marriages','Divorces'])
```
